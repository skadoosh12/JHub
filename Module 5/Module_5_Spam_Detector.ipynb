```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer as Tfidf, CountVectorizer as Cvec
from sklearn.model_selection import train_test_split as TTS
from sklearn.naive_bayes import MultinomialNB as MNB
from sklearn.svm import SVC
from wordcloud import WordCloud
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/JHub/Module 5/spam.csv", encoding = "ISO-8859-1")
df.head()
```

```python
df.drop(columns=["Unnamed: 2","Unnamed: 3", "Unnamed: 4"], inplace=True)
```

```python
#designate features and 
Y = df['v1']
X = df['v2']
```

```python
#Generate encoding objects for TFIDF and Count vectoriser and classifer objects

encs = [Tfidf(), Cvec(decode_error="ignore")]
models = [SVC(), MNB()]

for model in models:
  for enc in encs:
    #split data
    X_train, X_test, Y_train, Y_test = TTS(X, Y, test_size=0.33, random_state=1)

    #encode data
    obj = enc
    X_train = obj.fit_transform(X_train)

    #train model
    mod = model
    mod.fit(X_train, Y_train)

    #test model
    X_test = obj.transform(X_test)
    Y_predict = mod.predict(X_test)

    #Metrics
    cm = confusion_matrix(Y_test, Y_predict)
    ascore = accuracy_score(Y_test, Y_predict)

    #Output:
    print(model, enc, ascore)
```
