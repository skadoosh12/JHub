```python
word_net_lem = WordNetLemmatizer()
```

```python
stop_words = set(w.rstrip() for w in open("/content/drive/MyDrive/Colab Notebooks/JHub/Module 5/stopwords.txt"))
```

```python
positive_rev = BS(open("/content/drive/MyDrive/Colab Notebooks/JHub/Module 5/positive.review"))
positive_rev = positive_rev.findAll("review_text")

negative_rev = BS(open("/content/drive/MyDrive/Colab Notebooks/JHub/Module 5/negative.review"))
negative_rev = negative_rev.findAll("review_text")
```

```python
np.random.shuffle(positive_rev)
positive_rev = positive_rev[:len(negative_rev)]
```

```python
def my_tokeniser(s):
  s = s.lower()
  tokens = nltk.tokenize.word_tokenize(s)
  tokens = [t for t in tokens if len(t)>2]
  tokens = [word_net_lem.lemmatize(t) for t in tokens]
  tokens = [t for t in tokens if t not in stop_words]
  return tokens
```
